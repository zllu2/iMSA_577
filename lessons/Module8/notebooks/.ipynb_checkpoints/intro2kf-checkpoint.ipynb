## Simple Kalman Filter

We will simulate a factory's output as a noisy process, and use a Kalman filter to make predictions on the future output. This example is based on a Kalman filter example in the [scipy documentation][skfe].

First, we generate our signal along with the random noise. In this case, we expect on average 7.5 units to be produced per day. We sample the process over 100 days, and each day the fluctuations in the output are assumed to be normally distributed with a standard deviation of one-half unit. Next, we create our arrays, before specifying the Kalman parameters and initial states. Finally, we apply the Kalman filter to compute the predicted states at each observation, before displaying the observed data and predicted data. The results demonstrate the power of the Kalman filter to quickly zero in on the true factory output value.

-----
[skfe]: http://scipy-cookbook.readthedocs.io/items/KalmanFiltering.html

```python
# Initialization Date

num_iter = 100
sz = (num_iter)

truth = 7.5
sig = 0.5
obs = rng.normal(truth, sig, size = sz)
```

```python
# Working arrays
#
x_est = np.zeros(sz)
x_prior = np.zeros(sz)

err_est = np.zeros(sz)
err_prior = np.zeros(sz)

gain = np.zeros(sz)
```

```python
# Parameters controlling process

process_variance = 0.001
measure_variance = 0.01
```

```python
# Initial states

x_est[0] = 10.0
err_est[0] = 2.5
```

```python
# Perform Kalman filtering
for step in range(1, num_iter):
    
    # Update next expectation
    x_prior[step] = x_est[step - 1]
    err_prior[step] = err_est[step - 1] + process_variance
    
    # Update measurement
    gain[step] = err_prior[step] / (measure_variance + err_prior[step])
    x_est[step] = x_prior[step] + gain[step] * (obs[step] - x_prior[step])
    err_est[step] = (1 - gain[step]) * err_prior[step]
```

```python
# Sample points
x = np.linspace(0, num_iter - 1, num_iter)

# Plot Kalman Filtering 
fig, ax = plt.subplots(figsize=(12, 6))

ax.scatter(x, obs, marker='+', 
           c=sns.xkcd_rgb['denim blue'], alpha=0.75,
           label='Observed Data')

ax.plot(x, x_est, 
        c=sns.xkcd_rgb['dusty purple'], lw=2,
        label='Estimate')

ax.axhline(truth, linestyle='-.',
           color=sns.xkcd_rgb['dark green'], alpha=0.5,
           label='Truth')

ax.legend()

ax.set(title='Kalman Filtering',
      xlabel='Iteration', ylabel = 'Factory output')

sns.despine(offset=5, trim=True)
```

```python
# Plot convergence
fig, ax = plt.subplots(figsize=(12, 6))

# We skip first few, as they are big steps
ax.plot(x[2:], err_prior[2:],
        c=sns.xkcd_rgb['dusty purple'], lw=2,
        label='Estimate')

ax.set(title='Convergence',
      xlabel='Iteration', ylabel = 'Error in estimate')

sns.despine(offset=5, trim=True)
```
